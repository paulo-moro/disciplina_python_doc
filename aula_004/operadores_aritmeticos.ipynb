# Operadores

## Operadores Aritméticos

### Soma +
```
obj_1 + obj_2
 5 + 2 = 7
```

### Subtração -
```
obj_1 - obj_2
 5 - 2 = 3
```

### Multiplicação *
```
obj_1 * obj_2
 5 * 2 = 10
```

### Divisão /
```
obj_1 / obj_2
 5 / 2 = 2.5
 (Sempre retorna float)
```

### Resto %
```
obj_1 % obj_2
 5 % 2 = 1
 (sempre retorna um inteiro)
```

### Inteiro da divisão //
```
obj_1 // obj_2
 5 // 2 = 3
 (sempre retorna um inteiro)
```

### Exponenciação **
```
obj_1 ** obj_2
 5 ** 2 = 25
```
