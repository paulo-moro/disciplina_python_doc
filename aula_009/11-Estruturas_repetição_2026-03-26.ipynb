## Importe o .csv exportado na última aula e instancie um objeto da classe DataFrame.

Dos números de letras nos genes, quais são pares e quantos são impares?

```python

import pandas as pd
df_data = pd.read_csv('../storage/aula_004.csv')
df_data = df_data.drop('gene_name', axis=1)
df_data = df_data.T
impares = []
pares = []
for genes in df_data.index:
  if len(genes) % 2 == 0:
    pares.append(genes)
  else:
    impares.append(genes)
print(f"O numero de genes pares é {len(pares)} e são eles: {pares}")
print(f"O numero de genes impares é {len(impares)} e são eles: {impares}")
```
