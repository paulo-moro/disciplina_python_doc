## 01 – Obtenha os resultados e retorne o código em Python dos seguintes cálculos:

a) 2.0 + 3.0 / 5.0

b) (a + 10.2) - 5.0 * 2.0

c) a / b + 33.8

d) a / (b + c)

e) d * 2

```python
a = 2.0 + 3.0 / 5.0
print(f'Valor de a: {a}')
b = (a + 10.2) - 5.0 * 2.0
print(f'Valor de b: {b}')
c = a / b + 33.8
print(f'Valor de c: {c}')
d = a / (b + c)
print(f'Valor de d: {d}')
e = d * 2
print(f'Valor de e: {e}')
```

## 02 – Obtenha os resultados e retorne o código em Python dos seguintes cálculos:

a) [2.0, 3.0, 5.0] + [10, 20, 30]

b) lista_a * 2

c) lista_a + lista_b

d) lista_a / (lista_b * lista_c)

```python
lista_a = [2.0, 3.0, 5.0] + [10, 20, 30]
print(f'Valor de lista_a: {lista_a}')
lista_b = lista_a * 2
print(f'Valor de lista_b: {lista_b}')
lista_c = lista_a + lista_b
print(f'Valor de lista_c: {lista_c}')
lista_d = lista_a / (lista_b + lista_c)
print(f'Valor de lista_d: {lista_d}')
```

## 03 – Usando os objetos simples obtidos no exercício 01, instancie uma  lista. Instancie também uma segunda lista contendo a raiz quadrada dos valores obtidos nos objetos simples. Instancie uma terceira lista, concatenando as duas primeiras. Recupere a última lista.

```python
import math
lista_ex_1 = [a, b, c, d, e]
raiz_quadrada = [math.sqrt(x) for x in lista_ex_1]
lista_final = lista_ex_1 + raiz_quadrada
print(f'Raiz quadrada da lista ex 1: {raiz_quadrada}')
print(f'Lista final: {lista_final}')
```

## 04 – Duplique os valores contidos na lista final obtida no exercício anterior e converta para um array numpy. Remodele (reshape) o array, para que contenha 5 linhas e 4 colunas. Recupere o array.

```python
import numpy as np
lista_duplicada = lista_final * 2
np_array = np.array(lista_duplicada).reshape(5, 4)
print(f'Array numpy: \n{np_array}')
```

## 05 – Do array numpy obtenha uma lista contendo a média dos valores de cada uma das colunas.

```python
medias_colunas = np.mean(np_array, axis=0)

# Converter para lista
lista_medias = medias_colunas.tolist()

print("Médias de cada coluna:", lista_medias)
```

## 06 – O atributo shape pertence ao escopo dos objetos de classe numpy array. Ela retorna uma tupla onde o primeiro valor é o número de linhas e o segundo valor o de colunas contidas no array. Usando a informação do número total de valores presentes no numpy array do exercício 4 insira neste array a lista obtida no exercício 5. Recupere o array.
Obs.:  Atributos não usam () na chamada.

```python
print(f'Shape do array numpy: {np_array.shape}')
print(f'Lista medias: {lista_medias}')
array_expandido = np.insert(np_array, len(lista_medias)+1, lista_medias, axis=0)
array_expandido
print(array_expandido.shape)

```

## 07 – Remodele (reshape) o array do exercício anterior, para que contenha 4 colunas. Converta este objeto da classe array para DataFrame. Arredonde todos os valores para até 3 casas decimais. Recupere a DataFrame.

```python
import pandas as pd
reshape_exp = array_expandido.reshape(6, 4)
pd_reshaped = pd.DataFrame(reshape_exp.copy())
rounded_df = pd_reshaped.round(3)
print(f'DataFrame arredondado: \n{rounded_df}')

```

## 08 – A última instancia desta DataFrame deve conter as médias dos valores contidos por coluna, o que foi determinado nos exercícios anteriores. Instancie um objeto contendo as médias dos valores em cada linha. Insira este objeto como uma nova coluna na DataFrame. Recupere a DataFrame.

```python
medias_linhas = rounded_df.mean(1)
medias_linhas
rounded_df["media_linha"] = medias_linhas
print(f'DataFrame com média das linhas: \n{rounded_df}')
```

## 09 – Usando o método dos objetos de classe DataFrame, describe(), compare os resultados contidos na última linha e última coluna da DataFrame.

Obs.: O método dos objetos de classe DataFrame, transpose(), inverte linhas por colunas.

```python
column_describe = rounded_df.describe()
print(column_describe)
print('============================================')
row_describe = rounded_df.transpose().describe()
print(row_describe)
```

com o metodo describe foi possivel chegar nas mesmas médias de linhas e colunas bem interessante.

## 10 – Edite o nome das colunas e dos index, para indicar onde estão as médias. Qual a média das médias?

```python
custom_rows = ['line_1', 'line_2', 'line_3', 'line_4', 'line_5', 'media_colunas']
transposed_df = rounded_df.transpose()
transposed_df.columns = custom_rows
transposed_df
custom_columns = ['col_1', 'col_2', 'col_3', 'col_4', 'media_linha']
back_to_original = transposed_df.transpose()
back_to_original.columns = custom_columns
print(f'DataFrame final: \n{back_to_original}')
```
