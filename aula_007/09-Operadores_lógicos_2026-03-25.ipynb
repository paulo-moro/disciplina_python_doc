## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
print(p)
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
print(q)
r = 10 <= p / q + 33.8 < 100
print(r)
s = 5!= p /  (q+r)
print(s)
t = 33 == s * 2
print(t)
result = p or r and s or t
print('Para modificar o resultado basta modificar as comparações entre p e r para or e as comparações entre o resultado e t já que somente p e t são False, portanto havendo um valor verdadeiro tudo será verdadeiro ', result)
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

```python
import pandas as pd
df = pd.read_csv('../storage/aula_005.csv', index_col=0).round(3)
segunda_linha = df.iloc[1]
terceira_linha = df.iloc[2]
quarta_linha = df.iloc[3]
resultado = (segunda_linha is terceira_linha) or (terceira_linha is quarta_linha)
print("O resultado da condição é:", resultado)
```

## 04 – Usando operadores verifique se todos os valores contidos na primeira linha são maiores ou iguais a 10.

```python
primeira_linha = df.iloc[0]
print(primeira_linha)
resultado = (primeira_linha >= 10).all()

print("Todos os valores da primeira linha são >= 10?", resultado)
```

## 05 – Usando operadores verifique se todos os valores contidos na terceira coluna estão entre 0 inclusive e 40.

```python
terceira_coluna = df.iloc[:, 2]
resultado = ((terceira_coluna >= 0) & (terceira_coluna <= 40)).all()
print("Os valores da terceira coluna estão entre 0 e 40?", resultado)
```

## 06 – Usando operadores verifique se algum dos valores contidos na última linha está acima de 10.

```python
ultima_linha = df.iloc[-1]
result = (ultima_linha > 10).any()
result
```

## 08 – Recupere a DataFrame contendo apenas valores que estejam abaixo de 5 inclusive ou acima de 7.

## 09 – Converta as duas últimas colunas do objeto DataFrame para dois objetos da classe numpy array. Algum dos valores contidos no primeiro array estão entre 2 inclusive e 5 inclusive?

```python
print(df)
duas_ultimas_colunas = df.iloc[:, -2:]
array_1 = duas_ultimas_colunas.iloc[:, 0].to_numpy()
print(array_1)
array_2 = duas_ultimas_colunas.iloc[:, 1].to_numpy()
result = ((array_1 >= 2) & (array_1 <= 5)).any()
print("Algum valor do primeiro array está entre 2 e 5?", result)
```

## 10 – Os valores contidos nos dois arrays são superiores à -1 e inferiores à 12?

```python
primeiro_array_condicao = ((array_1 > -1) & (array_1 < 12))
segundo_array_condicao = ((array_2 > -1) & (array_2 < 12))
result = (primeiro_array_condicao & segundo_array_condicao).all()
print("Todos os valores dos arrays está estão -1 e 12 ?", result)
```
