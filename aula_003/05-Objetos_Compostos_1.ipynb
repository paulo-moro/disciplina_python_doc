## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Em cada uma das listas, você deve inserir o número de caracteres presentes no nome do gene, os dois primeiros caracteres do nome, quantas vezes aparece o caractere ‘a’ e se no nome aparecem os caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
lista_brca1 = [
  len(lista_de_nome_de_genes[0]), 
  lista_de_nome_de_genes[0][0:2], 
  lista_de_nome_de_genes[0].lower().count('a'),
  lista_de_nome_de_genes[0].lower().find('hu') > -1
]

lista_tp53 = [
  len(lista_de_nome_de_genes[1]), 
  lista_de_nome_de_genes[1][0:2], 
  lista_de_nome_de_genes[1].lower().count('a'),
  lista_de_nome_de_genes[1].lower().find('hu') > -1
]

lista_egfr = [
  len(lista_de_nome_de_genes[2]), 
  lista_de_nome_de_genes[2][0:2], 
  lista_de_nome_de_genes[2].lower().count('a'),
  lista_de_nome_de_genes[2].lower().find('hu') > -1
]

lista_vegfa = [
  len(lista_de_nome_de_genes[3]), 
  lista_de_nome_de_genes[3][0:2], 
  lista_de_nome_de_genes[3].lower().count('a'),
  lista_de_nome_de_genes[3].lower().find('hu') > -1
]

lista_apoe = [
  len(lista_de_nome_de_genes[4]), 
  lista_de_nome_de_genes[4][0:2], 
  lista_de_nome_de_genes[4].lower().count('a'),
  lista_de_nome_de_genes[4].lower().find('hu') > -1
]

print(lista_brca1, lista_tp53, lista_egfr, lista_vegfa, lista_apoe)
```
