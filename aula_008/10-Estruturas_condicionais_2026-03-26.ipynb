## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>

<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100

```python
p = 2.0 + 3.0 / 5.0 == 10
print(f'Valor de p: {p}')
q = (p +10.2) - 5.0 * 2.0 <= 2.9
print(f'Valor de q: {q}')
r = 10 <= p / q + 33.8 < 100
print(f'Valor de r: {r}')
s = 5 != p / (q + r)
print(f'Valor de s: {s}')
t = 33 == s * 2
print(f'Valor de t: {t}')
result  = p and r and s and t
if(result):
  print('O condicional p and r and s and t resultado é verdadeiro')
else:
  print('O condicional p and r and s and t resultado é falso')
```

```python
import numpy as np
matriz_exercicio_2_numpy = np.random.normal(10, 3, 5)
matriz_exercicio_2_numpy
count_up_ten = 0
for i in range(len(matriz_exercicio_2_numpy)):
  if matriz_exercicio_2_numpy[i] > 10:
    count_up_ten += 1
print(f'Número de elementos maiores que 10: {count_up_ten}')
```
